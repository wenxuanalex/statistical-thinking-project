```python
# Step 1 read the csv, find out what is missing/ not good enough in the csv
import pandas as pd
import numpy as np

df = pd.read_csv("airbnb sydney australia dataset.csv")

print(df.shape)
df.info()
df.describe(include='all')
```

```python
# Step 2 drop the columns that our group dont want and check the result
DROP_COLS = [
    "listing_url", "scrape_id", "last_scraped", "name", "source", "description", "picture_url", "host_name", "neighborhood_overview", "host_url",
    "host_name", "host_about", "host_thumbnail_url", "host_picture_url", "neighbourhood_group_cleansed", "minimum_minimum_nights", 
    "maximum_minimum_nights", "minimum_maximum_nights", "maximum_maximum_nights", "minimum_nights_avg_ntm", "maximum_nights_avg_ntm", 
    "calendar_updated", "calendar_last_scraped", "has_availability", "number_of_reviews_ltm", "number_of_reviews_l30d",
    "calculated_host_listings_count_entire_homes", "calculated_host_listings_count_private_rooms", "calculated_host_listings_count_shared_rooms", 
    "availability_30", "availability_60", "availability_90", "availability_eoy", "host_neighbourhood", "neighbourhood", "estimated_revenue_l365d", 
    "host_since", "host_location", "latitude", "longitude", "id", "host_id", "host_listings_count", "host_total_listings_count", "host_has_profile_pic",
    "host_verifications", "review_scores_accuracy", "review_scores_cleanliness", "review_scores_checkin", "review_scores_communication", "review_scores_location",
    "review_scores_value", "calculated_host_listings_count", "number_of_reviews_ly", "reviews_per_month", "neighbourhood_cleansed", "bathrooms_text",
    "beds", "availability_365", "property_type"
]

df = df.drop(columns=DROP_COLS, errors="ignore") # for this step, I ask chatgpt because when I run the second time, error appeared, so gpt tell me to add "errors = ingnore" to avoid that situation
print(df.shape)
df.info()
df.describe(include='all')
```

```python
# Step 3 change data type and modify data
currency_cols = ["price"]

for c in currency_cols:
    if c in df.columns:
        df[c] = df[c].astype(str).str.replace("$", "").str.replace(",", "")
        df[c] = df[c].astype(float)# for this step, I ask chatgpt because when I run the second time, error appeared, so gpt tell me to add "errors = coerce" to avoid that situation
```

```python
import pandas as pd

# Convert to datetime (month/day/year format)
df["first_review"] = pd.to_datetime(df["first_review"], format="%m/%d/%Y", errors="coerce")
df["last_review"]  = pd.to_datetime(df["last_review"],  format="%m/%d/%Y", errors="coerce")

# Define cutoff dates
cutoff_first = pd.Timestamp("2024-06-16")
cutoff_last  = pd.Timestamp("2024-06-17")

# Filter rows where:
df_filtered = df[
    (df["first_review"] <= cutoff_first) &
    (df["last_review"]  >= cutoff_last)
]

print("Original shape:", df.shape)
print("Filtered shape:", df_filtered.shape)
```

# Raw Data No Variables Transformed

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
# ------------------------
# OLS Model
# ------------------------
formula = """
occupancy_rate ~
host_response_rate + host_acceptance_rate +
price + minimum_nights + maximum_nights +
accommodates + bathrooms +
review_scores_rating + amenity_count +
C(host_response_time) + C(host_is_superhost) + C(host_identity_verified) +
C(room_type) + C(instant_bookable) + C(license)
"""

modelraw1 = smf.ols(formula=formula, data=df_trimmed).fit()
print(modelraw1.summary())

# ------------------------
# VIF Calculation
# ------------------------
# Use the design matrix from the fitted model
X = modelraw1.model.exog
vif_df = pd.DataFrame()
vif_df["Variable"] = modelraw1.model.exog_names
vif_df["VIF"] = [variance_inflation_factor(X, i) for i in range(X.shape[1])]

print("\nVariance Inflation Factors:")
print(vif_df)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Extract fitted values and residuals
fitted_vals1 = modelraw1.fittedvalues
residuals1 = modelraw1.resid

# Influence measures for standardized residuals
influence1 = modelraw1.get_influence()
std_resid1 = influence1.resid_studentized_internal
sqrt_std_resid1 = np.sqrt(np.abs(std_resid1))

# Create 1x3 diagnostic plots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 1. Residuals vs Fitted
axes[0].scatter(fitted_vals1, residuals1, alpha=0.5)
axes[0].axhline(y=0, color="red", linestyle="--")
axes[0].set_xlabel("Fitted values")
axes[0].set_ylabel("Residuals")
axes[0].set_title("Residuals vs Fitted")

# 2. QQ plot
sm.qqplot(residuals1, line="45", ax=axes[1])
axes[1].set_title("QQ Plot of Residuals")

# 3. Scale-Location plot
axes[2].scatter(fitted_vals1, sqrt_std_resid1, alpha=0.5)
axes[2].axhline(y=0, color="red", linestyle="--")
axes[2].set_xlabel("Fitted values")
axes[2].set_ylabel("âˆš|Standardized Residuals|")
axes[2].set_title("Scale-Location Plot")

plt.tight_layout()
plt.show()
```
